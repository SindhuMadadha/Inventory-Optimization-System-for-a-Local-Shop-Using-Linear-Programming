```python
import pulp

# Create optimization problem
model = pulp.LpProblem("Shop_Profit_Maximization", pulp.LpMaximize)

# Decision variables
x = pulp.LpVariable("Milk_Packets", lowBound=0, cat='Integer')
y = pulp.LpVariable("Bread_Packets", lowBound=0, cat='Integer')

# Objective function
model += 20*x + 15*y

# Constraints
model += x + y <= 100
model += 2*x + y <= 150

# Solve the problem
model.solve()

print("Milk Packets:", x.varValue)
print("Bread Packets:", y.varValue)
print("Maximum Profit:", pulp.value(model.objective))
```
